```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.iplt20.com/auction'
```

```python
data_url = requests.get(url)
```

```python
data_url
```

```python
#Getting the text from the url website
soup = BeautifulSoup(data_url.text,"lxml")
```

```python
soup
```

```python
#importing table from the data
table = soup.find("table",class_ = "ih-td-tab auction-tbl")
```

```python

print(table)
```

```python
title = table.find_all('th')
```

```python
title
```

```python
header = []
for i in title:
  name  = i.text
  header.append(name)
```

```python
header
```

```python
#creating dataframe
import pandas as pd
dff = pd.DataFrame(columns=header)
```

```python
dff
```

```python
rows = table.find_all('tr')
```

```python
row_data = []
for i in rows[1:]:
  data = i.find_all('td')
  row = [tr.text for tr in data]
  row_data.append(row)
  # l = len(df)
  # df.loc[l] = row
```

```python

len(row_data)
```

```python
#removing the empty list from the list
filtered_list = [item for item in row_data if item != []]
```

### How to append a list as a row to a Pandas DataFrame in Python?

```python



import pandas as pd

# List
Person = [ ['Satyam', 21, 'Patna' , 'India' ],
			['Anurag', 23, 'Delhi' , 'India' ],
			['Shubham', 27, 'Coimbatore' , 'India' ]]

#Create a DataFrame object
df = pd.DataFrame(Person,
				columns = ['Name' , 'Age', 'City' , 'Country'])

# display
display(df)
```

```python
# New list for append into df
list = ["Saurabh", 23, "Delhi", "india"]

# using loc methods
df.loc[len(df)] = list

# display
display(df)
```

```python
# Replace Values in a List using Lambda Function

# define list
l = ['Hardik', 'Rohit', 'Rahul', 'Virat', 'Pant']

# replace Pant with Ishan
l = list(map(lambda x: x.replace('Pant', 'Ishan'), l))

# print list
print(l)
```
